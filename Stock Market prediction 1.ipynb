```python
api_instance = upstox_client.HistoryApi()
instrument_key = 'NSE_INDEX|Nifty Bank' # str | Key of the instrument NSE_EQ|INE238A01034,NSE_INDEX|Nifty Bank

interval = 'day'
to_date ='2023-11-24'
from_date ='2021-11-08'
api_version = '2.0' # str | API Version Header
api_response = api_instance.get_historical_candle_data1(instrument_key,interval,to_date,from_date,api_version)
```

```python
df=df.dropna()
```

```python
df["Target"] = (df['Tomorrow'] > df['close']).astype(int)
```

```python
df = df.loc['2021-11-08' :].copy()
```

```python
df
```

```python
Training the model
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = df.iloc[:-100]
test = df.iloc[-100:]

predictors = ['close', 'volume', 'open', 'high', 'low']
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score
```

```python
preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds , index=test.index)
```

```python
preds
```

```python
precision_score(test['Target'],preds)
```
